### Line Plots

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
sns.lineplot(
    data = flights,
    x = 'year',
    y = 'passengers'
)
```

```python
sns.lineplot(
    data = flights,
    x = 'year',
    y = 'passengers',
    estimator = 'sum'
)
```

```python
sns.lineplot(
    data = flights,
    x = 'year',
    y = 'passengers',
    hue = 'month'
)
```

```python
taxis = sns.load_dataset('taxis', parse_dates = ['pickup','dropoff'])
```

```python
taxis
```

```python
sns.lineplot(
    data = taxis,
    x = taxis['pickup'].dt.hour,
    y = 'total',
    hue = 'payment',
    style = 'color',
    errorbar = ('ci', 1)
)
```

# RELPLOT (FIGURE level plotting)

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'sex')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'smoker', col = 'sex', row = 'time')
```

```python
sns.relplot(
    data = taxis,
    x = taxis['pickup'].dt.hour,
    y = 'total',
    kind = 'line',
    col = 'color',
    hue = 'payment'
)
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'sex', height = 2, aspect = 1)
```

### Histplot

```python
sns.histplot(
    data = tips,
    x = 'tip',
    hue = 'smoker'
)
```

```python
sns.histplot(
    data = tips,
    x = 'tip',
    hue = 'smoker',
    multiple = 'stack'
)
```

```python
sns.histplot(
    data = tips,
    x = 'tip',
    hue = 'smoker',
    multiple = 'dodge'
)
```

```python
sns.histplot(
    data = penguins,
    x = 'body_mass_g', 
    binwidth =200,
    hue = 'species',
    multiple = 'stack',
    element = 'step'
)
```

```python
sns.histplot(
    data = penguins,
    x = 'body_mass_g', 
    binwidth =200,
    hue = 'species',
    multiple = 'stack',
    element = 'step',
    kde = True
)
```

## Rugplot

```python
sns.rugplot(
    data = tips,
    x = 'tip',
    height = 0.2
)
```

```python
sns.kdeplot(data  = tips, x = 'total_bill')
sns.rugplot(data  = tips, x = 'total_bill', height = 0.1)
```

# displot (figure leveling plotting)

```python
sns.displot(
    kind = 'hist',
    data = penguins,
    x = 'body_mass_g',
    col = 'species',
    hue = 'sex',
    rug = True
)
```
